## Valores Nulos

#### Antes de aplicar cualquier transformación, analizamos la cantidad y el porcentaje de valores nulos de cada variable. Esta información nos permitirá decidir qué columnas conservar, tratar o eliminar durante el proceso de limpieza.

```python
nulls = pd.DataFrame({
    "Nulos": df.isnull().sum(),
    "Porcentaje (%)": round(df.isnull().mean() * 100, 2)
}).sort_values(by="Porcentaje (%)", ascending=False)

nulls
```

```python
nulls[nulls["Nulos"] > 0]
```
