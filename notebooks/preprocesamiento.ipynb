### Clasificación tamaño de las estaciones

#### Se crea una nueva columna que clasifica las estaciones según el total de kilómetros esquiables. Esta categorización facilitará la segmentación de los destinos y permitirá realizar comparaciones más representativas en el análisis.

```python
def clasificar_tamano(km):
    if km < 5:
        return "Muy pequeña"
    elif km < 25:
        return "Pequeña"
    elif km < 75:
        return "Mediana"
    elif km < 150:
        return "Grande"
    else:
        return "Muy grande"

df["Tamaño estación"] = df["Total Kms"].apply(clasificar_tamano)
```

### Clasificación del perfil recomendado

#### Se crea una nueva columna que clasifica cada estación según el tipo de pistas predominante. Esta información permitirá identificar el perfil de esquiador más adecuado para cada destino.

```python
df["% Pistas fáciles"] = (df["Easy"] / df["Total Kms"]) * 100
df["% Pistas intermedias"] = (df["Intermediate"] / df["Total Kms"]) * 100
df["% Pistas difíciles"] = (df["Difficult"] / df["Total Kms"]) * 100
```

```python
df.columns.tolist()
```

```python
def clasificar_perfil(row):
    
    porcentajes = {
        "Principiante": row["% Pistas fáciles"],
        "Intermedio": row["% Pistas intermedias"],
        "Avanzado": row["% Pistas difíciles"]
    }
    
    # Ordenar porcentajes de mayor a menor
    ordenados = sorted(porcentajes.items(), key=lambda x: x[1], reverse=True)

    # Si la diferencia entre el primero y el segundo es menor de 10 puntos,
    # consideramos la estación equilibrada
    if ordenados[0][1] - ordenados[1][1] < 10:
        return "Equilibrada"

    return ordenados[0][0]


df["Perfil recomendado"] = df.apply(clasificar_perfil, axis=1)
```

```python
df["Perfil recomendado"].value_counts()
```

```python
df[
    [
        "NameResort",
        "% Pistas fáciles",
        "% Pistas intermedias",
        "% Pistas difíciles",
        "Perfil recomendado"
    ]
].head(20)
```

```python
def clasificar_freeride(km):
    if pd.isna(km) or km == 0:
        return "Sin freeride"
    elif km <= 10:
        return "Freeride básico"
    elif km <= 30:
        return "Freeride avanzado"
    else:
        return "Para amantes del freeride"

df["Perfil freeride"] = df["Km Freeride"].apply(clasificar_freeride)
```
